```python
#Assessment 2:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Q 1: Develop a Line chart using the functionality of pandas to show how
automobile sales fluctuate from year to year.

```python

ds = pd.read_csv("/content/historical_automobile_sales.csv")
ds
```

```python
ds.columns
```

```python
sns.lineplot(x="Year",y="Automobile_Sales",data=ds,color= "green")
```

```python
ds ["Recession"] = ds["Recession"].map({1:"Yes",0:"No"})
```

```python
ds['Date'] = pd.to_datetime(ds['Date'])
```

Q 2: Plot different lines for categories of vehicle type and analyze the trend
to answer the question Is there a noticeable difference in sales trends
between different vehicle types during recession periods?

```python

ds['Date'] = pd.to_datetime(ds['Date'])
plt.figure(figsize=(14, 7))
sns.lineplot(data=ds,x='Date',y='Automobile_Sales',hue='Vehicle_Type',style='Recession')
plt.title("Automobile Sales Trends by Vehicle Type During Recession Periods")
plt.xlabel("Date")
plt.ylabel("Automobile Sales")
plt.legend(title='Vehicle Type / Recession')
plt.tight_layout()
plt.show()
```

Q 3: Use the functionality of Seaborn Library to create a visualization to compare
the sales trend per vehicle type for a recession period with a non- recession
period.

```python

plt.figure(figsize=(10,5))
sns.lineplot(x='Vehicle_Type',y='Automobile_Sales',hue='Recession',data=ds)
plt.title("Total Vehicle Sales: Recession VS No Recession")
```

```python
a = ds["Automobile_Sales"].sum()
```

Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period

```python

plt.figure(figsize=(10,5))
sns.barplot(x='Vehicle_Type',y='Automobile_Sales',hue='Recession',data=ds)
plt.title(f"Total Sales:-{a:.2f}")
```
